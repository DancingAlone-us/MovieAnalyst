```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mymoviedb.csv", engine = 'python')
df.head(5)
```

```python
df[(df["Genre"] == "Crime, Mystery, Thriller")]
```

```python
df["Primary Genre"] = df["Genre"].str.split(',').str[0]
df
```

```python
df[df["Release_Date"] == "None"]
```

```python
df['date_col'] = pd.to_datetime(df['Release_Date'], format='mixed')
df
```

```python
best_movies = (df["Vote_Average"] > 8.0) & (df["Vote_Count"] > 1000)
best_movies_in_netflix = df[best_movies]

print(best_movies_in_netflix[["Title", 'Vote_Average']].head(10))
```

```python
df
```

```python
plt.xlabel("Popularity")

plt.ylabel("Vote Average")



plt.scatter(df['Popularity'], df['Vote_Average'],c = "red")
plt.show()
```

```python
print(df["Original_Language"].value_counts().head(5))
```

```python
not_en = df[df["Original_Language"] != "en"]
major_language = not_en["Original_Language"].value_counts().head(5)
print(major_language)
```

```python
plt.bar(major_language.index, major_language.values, color = 'red')

plt.title("Language Diversity (ignoring en)")
plt.xlabel("Language")
plt.ylabel("No of Movies")

plt.show()
```

```python
top_genre = df.groupby("Primary Genre")['Popularity'].mean().sort_values(ascending = False)
best_genre = top_genre.head(5)
print(best_genre)
```

```python
plt.bar(best_genre.index, best_genre.values, color = "red")
plt.show()
```
